```python
import matplotlib.pyplot as plt
from operator import sub
import numpy as np
```

```python
objects = ('num_topics=3', 'num_topics=7', 'num_topics=16', 'Random')
y_pos = np.arange(len(objects))
min_r2 = [0.00608332767807,  0.00765454717902,  0.00763647763608 , 0.0074298026051]
max_r2 = [0.00947198128233, 0.00967813742898, 0.0116796008996, 0.00880778718308]
mean_r2 = [0.00732953004822, 0.00852765765745,  0.0092567600334, 0.00802891449442]
 
 
plt.bar(y_pos, mean_r2, color=['gray'], align='center',
        width=0.6, alpha=0.5, yerr = [map(sub, mean_r2, min_r2), map(sub, max_r2, mean_r2)],
        error_kw=dict(ecolor='gray', lw=2, capsize=5, capthick=2))
plt.xticks(y_pos, objects, fontsize=15)
plt.yticks(fontsize=22)
plt.ylabel('F1-score', fontsize=20)
plt.xlabel('Recommended method', fontsize=20)
plt.rcParams["figure.figsize"] = (10,6)
 
 
ax = plt.gca()
ax.set_ylim([0.006, 0.012])
plt.tight_layout()
plt.savefig("compare-recommend.pdf")
 
plt.show()
```
